# Naive Psuedo-Spectral Solver

STEP 1 : Import the solver class

```python
import sys
sys.path.insert(0, '../src')
import torch

from psuedo_spectral_solver_naive import ShallowWaterSolver as PSSolver
from visualization import animate_swe_on_box
```
